4. Mini Banking System with Transaction History
Design a Mini Banking System that simulates basic banking operations.
The system should support account creation, deposits, withdrawals, and fund transfers
between accounts. Each transaction must be recorded and retrievable as a mini statement.
Business rules include minimum balance enforcement, daily withdrawal limits, and
rejection of invalid transactions. Data persistence during runtime and menu-based
navigation are mandatory

```python
from datetime import date


accounts = {}

MIN_BALANCE = 500
DAILY_LIMIT = 10000


def add_transaction(acc_no, message):
    accounts[acc_no]["transactions"].append(message)

while True:
    print("\n===== MINI BANKING SYSTEM =====")
    print("1. Create Account")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Transfer Funds")
    print("5. Mini Statement")
    print("6. View Balance")




    print("7. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        acc_no = input("Enter new account number: ")
        if acc_no in accounts:
            print("Account already exists!")
        else:
            name = input("Enter account holder name: ")
            balance = float(input("Enter opening balance (min Rs.500): "))
            if balance < MIN_BALANCE:
                print("Minimum balance must be Rs.500")
            else:
                accounts[acc_no] = {
                    "name": name,
                    "balance": balance,
                    "transactions": [f"Account created with Rs.{balance}"],
                    "daily_withdrawn": 0,
                    "last_withdraw_date": date.today()
                }
                print("Account created successfully!")
    elif choice == "2":
        acc_no = input("Enter account number: ")
        if acc_no in accounts:
            amount = float(input("Enter deposit amount: "))
            if amount > 0:
                accounts[acc_no]["balance"] += amount
                add_transaction(acc_no, f"Deposited Rs.{amount}")
                print("Deposit successful!")
            else:
                print("Invalid amount.")
        else:
            print("Account not found.")
    elif choice == "3":
        acc_no = input("Enter account number: ")
        if acc_no in accounts:
            today = date.today()
            if accounts[acc_no]["last_withdraw_date"] != today:
                accounts[acc_no]["daily_withdrawn"] = 0
                accounts[acc_no]["last_withdraw_date"] = today

            amount = float(input("Enter withdrawal amount: "))

            if amount <= 0:
                print("Invalid amount.")
            elif accounts[acc_no]["daily_withdrawn"] + amount > DAILY_LIMIT:
                print("Daily withdrawal limit exceeded (Rs.10000).")
            elif accounts[acc_no]["balance"] - amount < MIN_BALANCE:
                print("Cannot withdraw. Minimum balance Rs.500 must be maintained.")
            else:
                accounts[acc_no]["balance"] -= amount
                accounts[acc_no]["daily_withdrawn"] += amount
                add_transaction(acc_no, f"Withdrew Rs.{amount}")
                print("Withdrawal successful!")
        else:
            print("Account not found.")

    elif choice == "4":
        sender = input("Enter your account number: ")
        receiver = input("Enter receiver account number: ")

        if sender in accounts and receiver in accounts:
            amount = float(input("Enter transfer amount: "))

            if amount <= 0:
                print("Invalid amount.")
            elif accounts[sender]["balance"] - amount < MIN_BALANCE:
                print("Transfer failed. Minimum balance rule violated.")
            else:
                accounts[sender]["balance"] -= amount
                accounts[receiver]["balance"] += amount

                add_transaction(sender, f"Transferred Rs.{amount} to {receiver}")
                add_transaction(receiver, f"Received Rs.{amount} from {sender}")

                print("Transfer successful!")
        else:
            print("Invalid account number(s).")
    elif choice == "5":
        acc_no = input("Enter account number: ")
        if acc_no in accounts:
            print("\n--- MINI STATEMENT ---")
            print("Name:", accounts[acc_no]["name"])
            for t in accounts[acc_no]["transactions"][-10:]:
                print("-", t)
        else:
            print("Account not found.")
    elif choice == "6":
        acc_no = input("Enter account number: ")
        if acc_no in accounts:
            print("Account Holder:", accounts[acc_no]["name"])
            print("Current Balance: Rs.", accounts[acc_no]["balance"])
        else:
            print("Account not found.")
    elif choice == "7":
        print("Thank you for using Mini Banking System.")
        break
    else:
        print("Invalid choice. Try again.")
```

**3. Campus Store Inventory and Billing System
Develop a console-based Inventory and Billing System for a campus store. The program should allow administrators to add products, update stock levels, and define
prices. Customers should be able to purchase multiple products in a single transaction. The system must automatically generate bills showing item-wise cost, subtotal, tax, discount (based on slabs), and final payable amount. Stock levels should update dynamically, and low-stock alerts should be displayed. Negative stock values must be prevented through validation logic. Use dictionaries for inventory management and lists for cart handling.**

```python
# Campus Store Inventory and Billing System

inventory = {}
LOW_STOCK_LIMIT = 5
TAX_RATE = 0.05

def add_product():
    name = input("Enter product name: ")
    price = float(input("Enter price: "))
    stock = int(input("Enter stock quantity: "))

    inventory[name] = [price, stock]
    print("Product added successfully!\n")

def update_stock():
    name = input("Enter product name: ")
    if name in inventory:
        stock = int(input("Enter new stock quantity: "))
        if stock >= 0:
            inventory[name][1] = stock
            print("Stock updated successfully!\n")
        else:
            print("Stock cannot be negative!\n")
    else:
        print("Product not found!\n")

def show_inventory():
    print("\nAvailable Products:")
    print("Product\t\t\tPrice\t\t\tStock")
    for name in inventory:
        price, stock = inventory[name]
        print(f"{name}\t\t\t{price}\t\t\t{stock}")

        if stock <= LOW_STOCK_LIMIT:
            print("** Low Stock Alert! **")
    print()

def calculate_discount(subtotal):
    if subtotal >= 1000:
        return subtotal * 0.10
    elif subtotal >= 500:
        return subtotal * 0.05
    else:
        return 0

def customer_purchase():
    cart = []
    subtotal = 0

    while True:
        show_inventory()
        name = input("Enter product name to buy (or 'done' to finish): ")

        if name.lower() == 'done':
            break

        if name not in inventory:
            print("Product not found!\n")
            continue

        qty = int(input("Enter quantity: "))

        price, stock = inventory[name]

        if qty <= 0:
            print("Quantity must be positive!\n")
        elif qty > stock:
            print("Not enough stock available!\n")
        else:
            cart.append([name, qty])
            inventory[name][1] -= qty
            subtotal += price * qty
            print("Added to cart!\n")

    if not cart:
        print("No items purchased.\n")
        return

    discount = calculate_discount(subtotal)
    tax = subtotal * TAX_RATE
    final_amount = subtotal + tax - discount

    print("\n------- BILL -------")
    print("Item\tQty\tPrice\tTotal")
    for item in cart:
        name, qty = item
        price = inventory[name][0]
        print(f"{name}\t{qty}\t{price}\t{price * qty}")

    print("--------------------")
    print("Subtotal:", subtotal)
    print("Tax (5%):", tax)
    print("Discount:", discount)
    print("Final Amount:", final_amount)
    print("--------------------\n")

def main_menu():
    while True:
        print("1. Add Product (Admin)")
        print("2. Update Stock (Admin)")
        print("3. Show Inventory")
        print("4. Customer Purchase")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_product()
        elif choice == '2':
            update_stock()
        elif choice == '3':
            show_inventory()
        elif choice == '4':
            customer_purchase()
        elif choice == '5':
            print("Thank you for using Campus Store System!")
            break
        else:
            print("Invalid choice!\n")

# Run the program
main_menu()
6
```
